```python
import pandas as pd
import os
import glob

# Caminho da pasta onde os arquivos Parquet estão sendo salvos
parquet_path = "../data/spark-jobs/output"

# Lista todos os arquivos parquet no diretório usando glob para pegar arquivos com padrão correto
parquet_files = glob.glob(os.path.join(parquet_path, "part-*.parquet"))

# Verificar se há arquivos para processar
if not parquet_files:
    print("Nenhum arquivo Parquet encontrado no diretório!")
else:
    print(f"Encontrados {len(parquet_files)} arquivos Parquet.")

    # Tenta ler cada arquivo individualmente para verificar se algum contém dados
    for file in parquet_files:
        print(f"\nTentando ler o arquivo: {file}")
        try:
            df_single = pd.read_parquet(file, engine="pyarrow")
            print(f"Conteúdo do arquivo {os.path.basename(file)}:")
            print(df_single.head())
            if df_single.empty:
                print("Arquivo está vazio (não contém linhas)")
        except Exception as e:
            print(f"Erro ao ler o arquivo: {e}")

    # Agora tenta concatenar todos
    try:
        df = pd.concat(
            [pd.read_parquet(f, engine="pyarrow") for f in parquet_files],
            ignore_index=True,
        )
        print("\nResultado da concatenação de todos os arquivos:")
        print(df.head())
    except Exception as e:
        print(f"Erro ao concatenar arquivos: {e}")
```
